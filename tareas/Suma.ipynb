```python
def ascending_coin(coin):
    """Returns the next ascending coin in order.
    >>> ascending_coin(1)
    5
    >>> ascending_coin(5)
    10
    >>> ascending_coin(10)
    25
    >>> ascending_coin(2) # Other values return None
    """
    if coin == 1:
        return 5
    elif coin == 5:
        return 10
    elif coin == 10:
        return 25

def descending_coin(coin):
    """Returns the next descending coin in order.
    >>> descending_coin(25)
    10
    >>> descending_coin(10)
    5
    >>> descending_coin(5)
    1
    >>> descending_coin(2) # Other values return None
    """
    if coin == 25:
        return 10
    elif coin == 10:
        return 5
    elif coin == 5:
        return 1

def count_coins(change):
    """Return the number of ways to make change using coins of value of 1, 5, 10, 25.
    >>> count_coins(15)
    6
    >>> count_coins(10)
    4
    >>> count_coins(20)
    9
    >>> count_coins(100) # How many ways to make change for a dollar?
    242
    >>> count_coins(200)
    1463
    >>> from construct_check import check
    >>> # ban iteration
    >>> check(HW_SOURCE_FILE, 'count_coins', ['While', 'For'])
    True
    """
    def contador(moneda, tamaño):

        print("Moneda y tamaño", moneda, tamaño)
        if moneda == 0:
            print("moneda", moneda)
            return 1
        elif moneda < 0:
            print("moneda", moneda)
            return 0
        elif tamaño == 25:
            return contador(moneda-tamaño, tamaño) 
        else:
            count_coin = contador(moneda-tamaño, tamaño)
            coin2 = contador(moneda, ascending_coin(tamaño))
            print("coin2", coin2)
            print("suma", count_coin + coin2)
            return count_coin + coin2
    return contador(change, 1) 

        
```
